# Variable Analyis - Univariate, Bivariate, Multivariatr

```python
numerical =["show_id","type","title","director","cast","country","date_added","release_year"]
numerical_axis = ["show_id","type","title","director","cast","country","date_added","release_year","rating","duration","listed_in","description"]
```

```python
list(zip(numerical,numerical_axis))
```

```python
title_font = {"family": "arial", "color":"darkblue","weight":"bold","size":15}
axis_font = {"family": "arial", "color":"darkred","weight":"bold","size":10}
for i,z in list(zip(numerical,numerical_axis)):
    plt.figure(figsize = (8,8), dpi = 90)
    sns.distplot(df["release_year"], hist_kws = dict(linewidth = 1, edgecolor = "g"), bins = 30)
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict= axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```
